```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
### GET FINAL RESULT ###

# Combine the data from the residential and commercial CSVs
res_df = pd.read_csv('combined_residential_results.csv')
com_df = pd.read_csv('combined_commercial_results.csv')
total_df = pd.concat([res_df,com_df], axis=0)

print("Number of samples:", len(total_df))

actual_site_eui = total_df['actual_site_eui'].tolist()
predict_site_eui = total_df['predict_site_eui'].tolist()

# Plot Results
plt.figure(figsize=(5,4))
plt.scatter(actual_site_eui,predict_site_eui)
plt.title("Random Forest on Commercial and Residential Buildings")
plt.xlabel("Actual site_eui Values")
plt.ylabel("Predicted site_eui Values")

total_score = metrics.r2_score(actual_site_eui, predict_site_eui)
print("Total score:", total_score)
```
